# Elbow curve Method

```python
dis =[]
from sklearn.cluster import KMeans
```

```python
k = range(2,10)
for i in k:
    model = KMeans(n_clusters = i)
    model.fit(data)
    dis.append(model.inertia_)
print(dis,end = " , ")
```

```python
distortion = []
for i in dis:
    distortion.append(round(i,2))
```

```python
distortion
```

```python
plt.figure(figsize =(15,7))
plt.plot(k,distortion,"r--o")
plt.xlabel("number of groups")
plt.ylabel('Distortion')
plt.title("Elbow_values")
plt.show()
```

```python
cluster = KMeans(n_clusters = 5)
model = cluster.fit(data)
labels = model.labels_
labels
```

```python
grouped_data = pd.DataFrame({"AnnualIncome(k$)":data[:,0],"Spending Score(1-100)":data[:,1],"Groups":labels})
```
